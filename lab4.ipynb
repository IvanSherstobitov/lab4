```python
import sys
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

# Изменим стандартный размер графиков matplotlib
plt.rcParams["figure.figsize"] = [6, 4]
```

```python
image = cv.imread('../images/lab4.jpg')
```

```python
# Использование библиотеки matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#%matplotlib inline позволяет выводить графики matplotlib в Jupyter
plt.imshow(image)
```

```python
image1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image1)
```

```python
thresholds = [100, 160]

prev_image = image

for tr in thresholds:
    ret, prev_image = cv.threshold(prev_image, tr, 255, cv.THRESH_BINARY)
plt.figure(figsize=(14, 10))
plt.imshow(prev_image, cmap='gray')
```

```python
cv.imwrite('ris.jpg', prev_image)
```
